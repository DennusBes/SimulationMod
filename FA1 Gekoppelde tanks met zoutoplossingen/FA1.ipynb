```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ik wil 4 uur aan tijd simuleren. 
# elk uur is 3600 seconden. 4 * 3600 = 14400 total seconds
# h is stepsize in minuten.    h=1/60   is dus 1 seconde.


def forward_euler(h,num_steps):
    # linker tank noem ik tank A, de rechter tank noem ik B
    # Dit zijn de waardes van de zoutoplossing (kg) per tank op moment 0.
    start_conc_a = float(0)
    start_conc_b = float(20)

    # hier maak ik lege array's en vul ik de waardes op moment 0 in.
    a = np.zeros(num_steps + 1)
    b = np.zeros(num_steps + 1)
    a[0] = start_conc_a
    b[0] = start_conc_b


    for step in range(num_steps):
        a[step + 1] = a[step] + (h*((0.2*6) + (b[step]*0.01) - (0.04*a[step]) - (0.03*a[step])))
        b[step + 1] = b[step] + (h*((0.03*a[step]) - (b[step]*0.01) - (b[step]*0.02)))
    return a,b


def plot_model(a,b):
    plt.plot(a)
    plt.plot(b)
    plt.legend(['Tank A', 'Tank B'])
    plt.xlabel('tijd (s)')
    plt.ylabel('zoutoplossing (kg)')
    plt.show()



# ik wil 4 uur aan tijd simuleren. 
# elk uur is 3600 seconden. 4 * 3600 = 14400 total seconds
# h is stepsize in minuten.    h=1/60   is dus 1 seconde.

a,b = forward_euler(1/60, 14400)
plot_model(a,b)
```
